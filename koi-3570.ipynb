```python
from astropy.io import fits
import glob as gl
import matplotlib.pyplot as plt
from mpl_style import lightbg
import numpy as np
```

```python
fnames = gl.glob('*.fits')
fnames = sorted(fnames)
hdu = fits.open(fnames[4])
print(fnames[5])
```

```python
time = hdu[1].data['TIME']
flux = hdu[1].data['SAP_FLUX']
len(flux)
```

```python
plt.style.use(lightbg)
plt.plot(time, flux, '.', markersize=1)
plt.xlim(1125, 1155)
plt.ylim(10000, 13000)
plt.xticks(np.arange(1125, 1160, 5))
plt.yticks(np.arange(10000, 14000, 1000))
plt.xlabel('Time (BJD - 2454833)')
plt.ylabel('Flux (e$^{-}$ per second)')
plt.tight_layout()
plt.savefig('koi-3570-lc.png')
```

```python
period = 3.649301
phase = (time - time[0]) / period % 1
```

```python
plt.plot(phase, flux, '.', markersize=1)
plt.xlim(0, 1)
plt.ylim(10000, 13000)
plt.xticks(np.arange(0, 1.2, 0.2))
plt.yticks(np.arange(10000, 14000, 1000))
plt.xlabel('Phase (Period = 3.649301 days)')
plt.ylabel('Flux (e$^{-}$ per second)')
plt.tight_layout()
plt.savefig('koi-3570-pflc.png')
```
